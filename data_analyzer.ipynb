```python
import numpy as np
import csv as np
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Parameters
main_folder = "O:\Projektarbeit\/2504"
input_folder = f"{main_folder}\Recorded"
output_folder = f"{main_folder}\Analyzed"
output_data = f"{main_folder}\Measured_values"
```

```python
# Read the text file into a DataFrame
df = pd.read_csv("O:\Projektarbeit\/1904\Testbench\BeAndFo10001.txt", sep='|', header=None, skiprows=1)

# Display the DataFrame
print(df)
```

```python
# Find the index where all three columns contain specific values
split_index_1 = df[(df[0] == -1) & (df[1] == -1) & (df[2] == -1)].index[0]
split_index_2 = df[(df[0] == -2) & (df[1] == -2) & (df[2] == -2)].index[0]
split_index_3 = df[(df[0] == -3) & (df[1] == -3) & (df[2] == -3)].index[0]

# Split the DataFrame into multiple parts
first_part = df.iloc[:split_index_1]
second_part = df.iloc[split_index_1+1:split_index_2]
third_part = df.iloc[split_index_2+1:split_index_3]
fourth_part = df.iloc[split_index_3+1:]

# Display the parts
print("First part:")
print(first_part)

print("\nSecond part:")
print(second_part)

print("\nThird part:")
print(third_part)

print("\nFourth part:")
print(fourth_part)
```

```python
df_drop = df.drop(index=[split_index_1, split_index_2, split_index_3])
# Rename the first column as "time"
# Rename the columns
df_drop.columns = ['Time', 'Force', '']

# Drop the third column
df_drop = df_drop.drop(columns=[''])
print(df_drop)
```

```python
# Plot
plt.figure(figsize=(10, 6))
plt.plot(df_drop['Time'], df_drop['Force'], marker='o', linestyle='-')
plt.title('Distance over Time')
plt.xlabel('Timestamp')
plt.ylabel('Distance (mm)')
plt.grid(True)
plt.show()
```
