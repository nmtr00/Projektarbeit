```python
import numpy as np
import csv 
import os
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
```

```python
# Read the text file into a DataFrame
df = pd.read_csv("1000N-0001.csv", sep = ";")

# Display the DataFrame
print(df)
```

```python


# Interpolate data for smoother plot
x_new = np.linspace(df['Timestamp'].min(), df['Timestamp'].max(), 300)
spl = make_interp_spline(df['Timestamp'], df['Distance (pixels)'], k=9)
y_smooth = spl(x_new)

# Plot
plt.figure(figsize=(10, 6))

# Original data plot
plt.plot(df['Timestamp'], df['Distance (pixels)'], label='Original', linestyle='-', color='blue')

# Smoothed data plot
plt.plot(x_new, y_smooth, label='Smoothed', linestyle='-', color='red')

plt.title('Distance over Time')
plt.xlabel('Timestamp')
plt.ylabel('Distance (mm)')
plt.legend()
plt.grid(True)
plt.show()
```
